In the baseline model, we use **TF-IDF** to transform article content into feature vectors. Then, we compute **cosine similarity** between these vectors to recommend similar articles. It's fast, simple, and effective!

For a deeper understanding of the text, we use **Large Language Models (LLMs)** like `DistilBERT` to generate embeddings. These embeddings capture richer semantic relationships in the content. We then compute cosine similarity on the embeddings for more context-aware recommendations.

### 5. Models Comparison
Lastly, we compare the performance of the TF-IDF model with the LLM embeddings. We'll weigh simplicity vs. contextual depth and consider which approach is more scalable and accurate for real-world use.

### 1. Data Inspection

```python
def inspect_data_comparative(train_df, dev_df, dataset_name):
    """
    Function to comprehensively inspect and compare train and dev datasets,
    including missing values, data types, statistics, and distribution of categorical variables.
    """
    print(f"\n{'='*60}")
    print(f" Dataset: {dataset_name}")
    print(f"{'='*60}\n")

    # Basic Info
    print(f"Train Dataset Shape: {train_df.shape}")
    print(f"Dev Dataset Shape: {dev_df.shape}")

    # Data Types and Missing Values
    def get_info_df(df, prefix):
        return pd.DataFrame({
            'Column': df.columns,
            'Data Type': df.dtypes,
            f'{prefix} Non-Null Count': df.notnull().sum(),
            f'{prefix} Null Count': df.isnull().sum(),
            f'{prefix} Null Percentage': (df.isnull().sum() / len(df)) * 100
        })

    train_info = get_info_df(train_df, 'Train')
    dev_info = get_info_df(dev_df, 'Dev')
    info_df = pd.merge(train_info, dev_info, on='Column')

    print("\nData Types and Missing Values:")
    print(info_df.to_string(index=False))

    # Basic Statistics
    print("\nBasic Statistics (Train):")
    print(train_df.describe(include='all').transpose().to_string())
    print("\nBasic Statistics (Dev):")
    print(dev_df.describe(include='all').transpose().to_string())

    # Categorical Variables Analysis
    excluded_columns = ['click_history', 'url', 'title_entities', 'abstract_entities', 'itemId']
    categorical_cols = train_df.select_dtypes(include=['object', 'category']).columns

    for col in categorical_cols:
        if col not in excluded_columns:
            print(f"\nUnique Values and Distribution for {col}:")

            train_counts = train_df[col].value_counts()
            dev_counts = dev_df[col].value_counts()

            print("Train Dataset:")
            print(train_counts.head(10).to_string())
            print(f"Number of unique values: {len(train_counts)}")

            print("\nDev Dataset:")
            print(dev_counts.head(10).to_string())
            print(f"Number of unique values: {len(dev_counts)}")

            # Plotting distribution
            plt.figure(figsize=(12, 6))

            # Get top 10 categories from both datasets
            top_categories = pd.concat([train_counts.head(10), dev_counts.head(10)]).index.unique()

            train_values = train_counts.reindex(top_categories).fillna(0)
            dev_values = dev_counts.reindex(top_categories).fillna(0)

            x = np.arange(len(top_categories))
            width = 0.35

            plt.bar(x - width/2, train_values, width, label='Train', color='blue', alpha=0.65)
            plt.bar(x + width/2, dev_values, width, label='Dev', color='orange', alpha=0.65)

            plt.title(f'Top categories in {col} - Train vs Dev')
            plt.xlabel(col)
            plt.ylabel('Count')
            plt.xticks(x, top_categories, rotation=45, ha='right')
            plt.legend()
            plt.tight_layout()
            plt.show()

    # Additional dataset-specific information
    if 'news' in dataset_name.lower():
        additional_info = pd.DataFrame({
            'Metric': ['Total Articles', 'Unique Categories', 'Unique Subcategories'],
            'Train Value': [
                len(train_df),
                train_df['category'].nunique(),
                train_df['subcategory'].nunique()
            ],
            'Dev Value': [
                len(dev_df),
                dev_df['category'].nunique(),
                dev_df['subcategory'].nunique()
            ]
        })
    elif 'behaviors' in dataset_name.lower():
        train_df['timestamp'] = pd.to_datetime(train_df['timestamp'], format='%m/%d/%Y %I:%M:%S %p')
        dev_df['timestamp'] = pd.to_datetime(dev_df['timestamp'], format='%m/%d/%Y %I:%M:%S %p')
        additional_info = pd.DataFrame({
            'Metric': ['Total Impressions', 'Unique Users', 'Date Range'],
            'Train Value': [
                len(train_df),
                train_df['userId'].nunique(),
                f"{train_df['timestamp'].min()} to {train_df['timestamp'].max()}"
            ],
            'Dev Value': [
                len(dev_df),
                dev_df['userId'].nunique(),
                f"{dev_df['timestamp'].min()} to {dev_df['timestamp'].max()}"
            ]
        })
    else:
        additional_info = pd.DataFrame()

    print("\nAdditional Information:")
    print(additional_info.to_string(index=False))

    return {
        'info_df': info_df,
        'additional_info': additional_info
    }

# Run inspections
news_result = inspect_data_comparative(train_news, dev_news, "News Dataset")
behaviors_result = inspect_data_comparative(train_behaviors, dev_behaviors, "Behaviors Dataset")
```

```python
def analyze_categorical_distribution(train_df, dev_df, columns, dataset_name):
    """
    Function to analyze and compare the distribution of categorical variables
    between train and dev datasets.
    """
    for column in columns:
        if train_df[column].dtype == 'object' and dev_df[column].dtype == 'object':
            print(f"\n{'='*60}")
            print(f"Distribution of {column} in {dataset_name}:")
            print(f"{'='*60}")

            # Get value counts
            train_counts = train_df[column].value_counts()
            dev_counts = dev_df[column].value_counts()

            # Combine and sort by total count
            combined_counts = pd.concat([train_counts, dev_counts], axis=1, keys=['Train', 'Dev'])
            combined_counts['Total'] = combined_counts.sum(axis=1)
            combined_counts = combined_counts.sort_values('Total', ascending=False)

            # Display top 10 categories
            print("\nTop 10 categories:")
            print(combined_counts.head(10).to_string())

            # Display summary statistics
            print(f"\nUnique values in Train: {train_df[column].nunique()}")
            print(f"Unique values in Dev: {dev_df[column].nunique()}")

            # Plotting
            plt.figure(figsize=(12, 6))

            # Get top 10 categories
            top_categories = combined_counts.head(10).index

            # Prepare data for plotting
            train_data = train_counts.reindex(top_categories).fillna(0)
            dev_data = dev_counts.reindex(top_categories).fillna(0)

            # Create bar plot
            x = range(len(top_categories))
            width = 0.35

            plt.bar([i - width/2 for i in x], train_data, width, label='Train', color='blue', alpha=0.65)
            plt.bar([i + width/2 for i in x], dev_data, width, label='Dev', color='orange', alpha=0.65)

            plt.title(f'Top 10 Categories in {column} - {dataset_name}')
            plt.xlabel('Category')
            plt.ylabel('Count')
            plt.xticks(x, top_categories, rotation=45, ha='right')
            plt.legend()
            plt.tight_layout()
            plt.show()

        else:
            print(f"\n{column} is not a categorical column in both datasets.")

# Analyze unique values and categorical distribution for news datasets
analyze_categorical_distribution(train_news, dev_news, ['category', 'subcategory'], "News Dataset")

# Analyze unique values and categorical distribution for behaviors datasets
analyze_categorical_distribution(train_behaviors, dev_behaviors, ['userId'], "Behaviors Dataset")
```

### 2. Behavioral Analysis

```python
def analyze_user_interactions(train_df, dev_df, dataset_name):
    """
    Analyze how often users click on certain articles in both train and dev datasets.
    Handles NaN values in click_history.
    """
    print(f"\n{'='*60}")
    print(f"User Interaction Analysis for {dataset_name}")
    print(f"{'='*60}")

    def process_clicks(df):
        # Filter out NaN values and convert to string
        click_histories = df['click_history'].dropna().astype(str)
        all_clicks = [item for sublist in click_histories.str.split() for item in sublist if item]
        click_counts = Counter(all_clicks)
        return pd.Series(click_counts).sort_values(ascending=False)

    train_clicks = process_clicks(train_df)
    dev_clicks = process_clicks(dev_df)

    print("\nTop 10 Most Clicked Articles:")
    combined_clicks = pd.DataFrame({'Train': train_clicks, 'Dev': dev_clicks}).fillna(0)
    combined_clicks['Total'] = combined_clicks['Train'] + combined_clicks['Dev']
    combined_clicks = combined_clicks.sort_values('Total', ascending=False)
    print(combined_clicks.head(10).to_string())

    print(f"\nTotal Unique Articles Clicked in Train: {len(train_clicks)}")
    print(f"Total Unique Articles Clicked in Dev: {len(dev_clicks)}")

    # Plotting
    plt.figure(figsize=(12, 6))
    top_articles = combined_clicks.head(10).index

    plt.bar(range(10), combined_clicks['Train'].head(10), label='Train', alpha=0.65, color='blue')
    plt.bar(range(10), combined_clicks['Dev'].head(10), bottom=combined_clicks['Train'].head(10),
            label='Dev', alpha=0.65, color='orange')

    plt.title(f'Top 10 Clicked Articles - {dataset_name}')
    plt.xlabel('Article ID')
    plt.ylabel('Number of Clicks')
    plt.xticks(range(10), top_articles, rotation=45, ha='right')
    plt.legend()
    plt.tight_layout()
    plt.show()

    # Additional analysis: Users with no click history
    train_no_clicks = train_df['click_history'].isna().sum()
    dev_no_clicks = dev_df['click_history'].isna().sum()
    print(f"\nUsers with no click history in Train: {train_no_clicks} ({train_no_clicks/len(train_df)*100:.2f}%)")
    print(f"Users with no click history in Dev: {dev_no_clicks} ({dev_no_clicks/len(dev_df)*100:.2f}%)")

# Assuming train_behaviors and dev_behaviors are already loaded
analyze_user_interactions(train_behaviors, dev_behaviors, "Behaviors Dataset")
```

### 3. Content Analysis

```python
def preprocess_text(text):
    """Clean and preprocess text data."""
    if pd.isna(text):
        return ""
    # Convert to lowercase and remove special characters
    text = re.sub(r'[^a-zA-Z\s]', '', str(text).lower())
    # Tokenize
    tokens = nltk.word_tokenize(text)
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    tokens = [t for t in tokens if t not in stop_words and len(t) > 2]
    return ' '.join(tokens)

def analyze_frequent_words(train_df, dev_df, column, dataset_name, top_n=20):
    """Analyze and compare most frequent words in train and dev datasets."""
    print(f"\n{'='*60}")
    print(f"Frequent Words Analysis for {column} in {dataset_name}")
    print(f"{'='*60}")

    def get_top_words(df, col):
        text = ' '.join(df[col].fillna('').apply(preprocess_text))
        words = nltk.word_tokenize(text)
        return Counter(words).most_common(top_n)

    train_top = dict(get_top_words(train_df, column))
    dev_top = dict(get_top_words(dev_df, column))

    all_words = list(set(train_top.keys()) | set(dev_top.keys()))
    combined = pd.DataFrame({
        'Train': [train_top.get(word, 0) for word in all_words],
        'Dev': [dev_top.get(word, 0) for word in all_words]
    }, index=all_words)
    combined['Total'] = combined['Train'] + combined['Dev']
    combined = combined.sort_values('Total', ascending=False).head(top_n)

    print(f"\nTop {top_n} Most Frequent Words:")
    print(combined.to_string())

    # Plotting
    plt.figure(figsize=(12, 6))
    combined[['Train', 'Dev']].plot(kind='bar', color=['blue', 'orange'], alpha=0.7, ax=plt.gca())
    plt.title(f'Top {top_n} Words in {column} - {dataset_name}')
    plt.xlabel('Words')
    plt.ylabel('Frequency')
    plt.legend(loc='upper right')
    plt.xticks(rotation=45, ha='right')
    plt.tight_layout()
    plt.show()

    # Word Cloud
    wc = WordCloud(width=800, height=400, background_color='white').generate_from_frequencies(dict(combined['Total']))
    plt.figure(figsize=(10, 5))
    plt.imshow(wc, interpolation='bilinear')
    plt.axis('off')
    plt.title(f'Word Cloud for {column} - {dataset_name}')
    plt.show()

def extract_keywords_and_topics(train_df, dev_df, column, dataset_name, n_keywords=10):
    """Extract keywords and categorize topics using TF-IDF."""
    print(f"\n{'='*60}")
    print(f"Keyword Extraction and Topic Categorization for {column} in {dataset_name}")
    print(f"{'='*60}")

    tfidf = TfidfVectorizer(max_features=1000, stop_words='english', ngram_range=(1,2))

    def get_top_tfidf_words(df, col):
        tfidf_matrix = tfidf.fit_transform(df[col].fillna('').apply(preprocess_text))
        feature_names = tfidf.get_feature_names_out()
        tfidf_scores = tfidf_matrix.sum(axis=0).A1
        word_scores = list(zip(feature_names, tfidf_scores))
        return sorted(word_scores, key=lambda x: x[1], reverse=True)[:n_keywords]

    train_keywords = get_top_tfidf_words(train_df, column)
    dev_keywords = get_top_tfidf_words(dev_df, column)

    print("\nTop Keywords (TF-IDF) in Train Dataset:")
    for word, score in train_keywords:
        print(f"{word}: {score:.4f}")

    print("\nTop Keywords (TF-IDF) in Dev Dataset:")
    for word, score in dev_keywords:
        print(f"{word}: {score:.4f}")

    # Simple topic categorization based on top keywords
    def categorize_topics(keywords):
        topics = []
        for word, _ in keywords:
            if 'politic' in word or 'government' in word:
                topics.append('Politics')
            elif 'econom' in word or 'financ' in word or 'business' in word:
                topics.append('Economy')
            elif 'health' in word or 'medic' in word:
                topics.append('Health')
            elif 'tech' in word or 'scienc' in word:
                topics.append('Technology/Science')
            elif 'sport' in word:
                topics.append('Sports')
            else:
                topics.append('Other')
        return Counter(topics).most_common()

    train_topics = categorize_topics(train_keywords)
    dev_topics = categorize_topics(dev_keywords)

    print("\nPredominant Topics in Train Dataset:")
    for topic, count in train_topics:
        print(f"{topic}: {count}")

    print("\nPredominant Topics in Dev Dataset:")
    for topic, count in dev_topics:
        print(f"{topic}: {count}")

# Assuming train_news and dev_news are already loaded
analyze_frequent_words(train_news, dev_news, 'title', "News Dataset")
analyze_frequent_words(train_news, dev_news, 'abstract', "News Dataset")
extract_keywords_and_topics(train_news, dev_news, 'title', "News Dataset")
extract_keywords_and_topics(train_news, dev_news, 'abstract', "News Dataset")
```

# 4. Modeling

### 1. Preprocessing Data

```python
# Preprocessing Function
def preprocess_data(news_df: pd.DataFrame) -> pd.DataFrame:
    """Preprocess news data by combining title and abstract, cleaning text, and normalizing."""
    # Fill missing values in 'title' and 'abstract'
    news_df['title'] = news_df['title'].fillna('')
    news_df['abstract'] = news_df['abstract'].fillna('')
    
    # Combine 'title' and 'abstract' into 'content'
    news_df['content'] = news_df['title'] + " " + news_df['abstract']
    
    # Text cleaning: remove special characters, numbers, and convert to lowercase
    news_df['content'] = news_df['content'].apply(lambda x: re.sub(r'[^a-zA-Z\s]', '', x).lower())
    
    # Remove stopwords
    stop_words = set(stopwords.words('english'))
    news_df['content'] = news_df['content'].apply(lambda x: ' '.join([word for word in x.split() if word not in stop_words]))
    
    # Perform stemming or lemmatization
    lemmatizer = WordNetLemmatizer()
    news_df['content'] = news_df['content'].apply(lambda x: ' '.join([lemmatizer.lemmatize(word) for word in x.split()]))
    
    return news_df

# Preprocess the news datasets
train_news = preprocess_data(train_news)
dev_news = preprocess_data(dev_news)

# Serialize preprocessed news data to CSV
train_news.to_csv("train_news.csv", index=False)
dev_news.to_csv("dev_news.csv", index=False)
print("Preprocessed news data saved.")

# Display preprocessed news data
display(train_news.head())
```

### 2. TF-IDF Vectorization

```python
# Create and Save TF-IDF Vectors
def create_tfidf_vectors(train_content: List[str], dev_content: List[str], max_features: int = 10000, ngram_range: Tuple[int, int] = (1, 2)) -> Tuple[np.ndarray, np.ndarray, TfidfVectorizer]:
    """Create TF-IDF vectors for train and dev content with enhanced parameters."""
    vectorizer = TfidfVectorizer(stop_words='english', max_features=max_features, ngram_range=ngram_range)
    train_tfidf = vectorizer.fit_transform(train_content)
    dev_tfidf = vectorizer.transform(dev_content)
    return train_tfidf, dev_tfidf, vectorizer

# Generate TF-IDF vectors with enhanced parameters
train_tfidf, dev_tfidf, tfidf_vectorizer = create_tfidf_vectors(train_news['content'], dev_news['content'])

# Serialize the TF-IDF vectors and vectorizer using joblib
joblib.dump(train_tfidf, "train_tfidf_vectors.joblib")
joblib.dump(dev_tfidf, "dev_tfidf_vectors.joblib")
joblib.dump(tfidf_vectorizer, "tfidf_vectorizer.joblib")
print("TF-IDF model and vectors saved.")

# Display shape of TF-IDF vectors
print(f"Train TF-IDF shape: {train_tfidf.shape}")
print(f"Dev TF-IDF shape: {dev_tfidf.shape}")
```

### 3. LLM Embeddings

```python
def get_llm_embeddings(texts: List[str], model_name: str = 'distilbert-base-nli-stsb-mean-tokens', batch_size: int = 32) -> np.ndarray:
    """Generate embeddings using a pre-trained language model with enhanced performance options."""
    model = SentenceTransformer(model_name)
    embeddings = model.encode(texts, batch_size=batch_size, show_progress_bar=True, normalize_embeddings=True)
    return embeddings

# Generate LLM embeddings for the dev dataset
dev_embeddings = get_llm_embeddings(dev_news['content'].tolist())

# Serialize the embeddings using joblib
joblib.dump(dev_embeddings, "dev_llm_embeddings.joblib")
print("LLM embeddings saved.")

# Display the shape of LLM embeddings
print(f"Dev Embeddings shape: {dev_embeddings.shape}")
```

### 4. Cosine Similarity Calculation

```python
def calculate_cosine_similarity(embeddings: np.ndarray, batch_size: int = 1000) -> np.ndarray:
    """Calculate cosine similarity matrix from embeddings in batches to optimize memory usage."""
    n = embeddings.shape[0]
    cosine_sim_matrix = np.zeros((n, n))
    for start_idx in range(0, n, batch_size):
        end_idx = min(start_idx + batch_size, n)
        cosine_sim_matrix[start_idx:end_idx] = cosine_similarity(embeddings[start_idx:end_idx], embeddings)
    return cosine_sim_matrix

# Calculate cosine similarity matrices for TF-IDF and LLM embeddings
cosine_sim_tfidf = calculate_cosine_similarity(dev_tfidf)
cosine_sim_llm = calculate_cosine_similarity(dev_embeddings)

# Serialize the cosine similarity matrices using joblib
joblib.dump(cosine_sim_tfidf, "cosine_sim_tfidf.joblib")
joblib.dump(cosine_sim_llm, "cosine_sim_llm.joblib")
print("Cosine similarity matrices saved.")

# Display shapes of cosine similarity matrices
print(f"Cosine Similarity TF-IDF shape: {cosine_sim_tfidf.shape}")
print(f"Cosine Similarity LLM shape: {cosine_sim_llm.shape}")
```

### 5. Get Similar Articles

```python
def load_precomputed_data() -> Tuple:
    """
    Load precomputed models and data required for generating recommendations.

    Returns:
        tuple: Loaded TF-IDF vectorizer, TF-IDF vectors, LLM embeddings, cosine similarity matrices, and news dataframe.
    """
    try:
        vectorizer = joblib.load("tfidf_vectorizer.joblib")
        dev_tfidf = joblib.load("dev_tfidf_vectors.joblib")
        dev_embeddings = joblib.load("dev_llm_embeddings.joblib")
        cosine_sim_tfidf = joblib.load("cosine_sim_tfidf.joblib")
        cosine_sim_llm = joblib.load("cosine_sim_llm.joblib")
        dev_news = pd.read_csv("dev_news.csv")
        return vectorizer, dev_tfidf, dev_embeddings, cosine_sim_tfidf, cosine_sim_llm, dev_news
    except FileNotFoundError as e:
        raise FileNotFoundError(f"Error loading data: {e}")
    except Exception as e:
        raise RuntimeError(f"Unexpected error: {e}")
```

```python
def get_similar_articles(article_id: str, news_df: pd.DataFrame, cosine_sim: np.ndarray, num_recommendations: int = 5) -> pd.DataFrame:
    """
    Get similar articles based on cosine similarity.

    Args:
        article_id (str): ID of the article to find similar articles for.
        news_df (pd.DataFrame): DataFrame containing news articles.
        cosine_sim (np.ndarray): Precomputed cosine similarity matrix.
        num_recommendations (int, optional): Number of recommendations to return. Defaults to 5.

    Returns:
        pd.DataFrame: DataFrame containing the recommended articles with similarity scores.
    """
    if article_id not in news_df['itemId'].values:
        raise ValueError(f"Article ID {article_id} not found in the dataset.")

    idx = news_df[news_df['itemId'] == article_id].index[0]
    sim_scores = list(enumerate(cosine_sim[idx]))
    sim_scores = sorted(sim_scores, key=lambda x: x[1], reverse=True)
    top_similar_indices = [i for i, _ in sim_scores[1:num_recommendations + 1]]

    similar_articles = news_df.iloc[top_similar_indices].copy()
    similar_articles['similarity_score'] = [cosine_sim[idx][i] for i in top_similar_indices]

    return similar_articles[['itemId', 'title', 'abstract', 'category', 'subcategory', 'similarity_score']]

def display_recommendations(article_id: str, news_df: pd.DataFrame, cosine_sim_llm: np.ndarray, cosine_sim_tfidf: np.ndarray, num_recommendations: int = 5) -> Tuple[pd.DataFrame, pd.DataFrame]:
    """
    Display similar articles using both TF-IDF and LLM-based similarities for comparison.

    Args:
        article_id (str): ID of the article to find similar articles for.
        news_df (pd.DataFrame): DataFrame containing news articles.
        cosine_sim_llm (np.ndarray): Cosine similarity matrix from LLM embeddings.
        cosine_sim_tfidf (np.ndarray): Cosine similarity matrix from TF-IDF vectors.
        num_recommendations (int, optional): Number of recommendations to return. Defaults to 5.

    Returns:
        tuple: DataFrames containing LLM-based and TF-IDF-based recommendations.
    """
    similar_articles_llm = get_similar_articles(article_id, news_df, cosine_sim_llm, num_recommendations)
    similar_articles_tfidf = get_similar_articles(article_id, news_df, cosine_sim_tfidf, num_recommendations)

    return similar_articles_llm, similar_articles_tfidf

# Load precomputed data
try:
    vectorizer, dev_tfidf, dev_embeddings, cosine_sim_tfidf, cosine_sim_llm, dev_news = load_precomputed_data()
    
    # Example usage
    article_id = dev_news['itemId'].iloc[0]  # Get the first article ID as an example
    similar_articles_llm, similar_articles_tfidf = display_recommendations(article_id, dev_news, cosine_sim_llm, cosine_sim_tfidf)
    
    # Display the recommendations in DataFrame format
    print("\nRecommendations using LLM-based similarity:\n")
    display(similar_articles_llm)
    print("\nRecommendations using TF-IDF-based similarity:\n")
    display(similar_articles_tfidf)
except Exception as e:
    print(f"An error occurred: {e}")
```

### 6. Get User Recommendations 

```python
def get_user_recommendations(user_id: str, behaviors_df: pd.DataFrame, news_df: pd.DataFrame, cosine_sim: np.ndarray, num_recommendations: int = 5) -> pd.DataFrame:
    """Get personalized recommendations for a user based on their click history using cosine similarity."""

    # Step 1: Extract user click history
    try:
        user_history = behaviors_df[behaviors_df['userId'] == user_id]['click_history'].iloc[0].split()  # List of clicked article IDs
    except IndexError:
        # If user ID is not found, return an empty DataFrame
        return pd.DataFrame(columns=['itemId', 'title', 'abstract', 'category', 'subcategory', 'similarity_score'])

    user_indices = [news_df[news_df['itemId'] == item_id].index[0] for item_id in user_history if item_id in news_df['itemId'].values]

    # Step 2: Handle case where user has no click history
    if not user_indices:
        # Return the most popular articles (if popularity data available) or random recommendations
        return news_df.sample(num_recommendations)[['itemId', 'title', 'abstract', 'category', 'subcategory']].assign(similarity_score=np.nan)

    # Step 3: Calculate similarity scores for articles
    sim_scores = defaultdict(float)
    for idx in user_indices:
        similar_articles = list(enumerate(cosine_sim[idx]))
        for i, score in similar_articles:
            if i != idx and news_df.iloc[i]['itemId'] not in user_history:  # Exclude self-similarity and already clicked articles
                sim_scores[i] = max(sim_scores[i], score)  # Use the maximum similarity score to avoid inflated values

    # Step 4: Sort and select top recommendations excluding user's clicked articles
    sorted_scores = sorted(sim_scores.items(), key=lambda x: x[1], reverse=True)
    recommended_indices = [i for i, score in sorted_scores][:num_recommendations]

    # Step 5: Create a DataFrame for the recommended articles
    recommendations = pd.DataFrame({
        'itemId': [news_df.iloc[i]['itemId'] for i in recommended_indices],
        'title': [news_df.iloc[i]['title'] for i in recommended_indices],
        'abstract': [news_df.iloc[i]['abstract'] for i in recommended_indices],
        'category': [news_df.iloc[i]['category'] for i in recommended_indices],
        'subcategory': [news_df.iloc[i]['subcategory'] for i in recommended_indices],
        'similarity_score': [sim_scores[i] for i in recommended_indices]
    })

    return recommendations

# Example usage: Get recommendations for a specific user
user_id = dev_behaviors['userId'].iloc[0]
user_recommendations = get_user_recommendations(user_id, dev_behaviors, dev_news, cosine_sim_llm)
display(user_recommendations)
```

```python
def calculate_mrr(relevance: list) -> float:
    """Calculate Mean Reciprocal Rank (MRR)."""
    for rank, rel in enumerate(relevance, start=1):
        if rel == 1:
            return 1.0 / rank
    return 0.0
```

### 6. Recommendation System Evaluation

```python
def evaluate_average_metrics_for_users(user_ids: List[str], recommender_func, news_df: pd.DataFrame, behaviors_df: pd.DataFrame, cosine_sim: np.ndarray, k: int = 10) -> Dict[str, float]:
    """Evaluate average metrics across a list of users."""
    results = []
    
    for user_id in user_ids:
        user_behavior = behaviors_df[behaviors_df['userId'] == user_id]
        if user_behavior.empty or pd.isna(user_behavior['click_history'].iloc[0]):
            continue

        click_history = user_behavior['click_history'].iloc[0].split()
        if not click_history:
            continue

        try:
            # Use the latest clicked article's index if it is a valid integer index
            latest_click = click_history[-1]
            if latest_click not in news_df['itemId'].values:
                continue

            recommended_articles = recommender_func(latest_click, news_df, cosine_sim, num_recommendations=k)
            recommended_ids = recommended_articles['itemId'].tolist()
            relevance = [1 if item in click_history else 0 for item in recommended_ids]

            # Calculate AUC
            true_labels = [1 if item in click_history else 0 for item in news_df['itemId']]
            predicted_scores = [cosine_sim[news_df[news_df['itemId'] == latest_click].index[0], news_df[news_df['itemId'] == item].index[0]] if item in recommended_ids else 0 for item in news_df['itemId']]
            auc_score = roc_auc_score(true_labels, predicted_scores) if len(set(true_labels)) > 1 else 0.0

            # Calculate MRR
            mrr_score = calculate_mrr(relevance)

            # Calculate nDCG@5 and nDCG@10
            ideal_relevance = sorted(relevance, reverse=True)
            ndcg_5 = ndcg_score([ideal_relevance[:5]], [relevance[:5]]) if sum(relevance[:5]) > 0 else 0.0
            ndcg_10 = ndcg_score([ideal_relevance[:10]], [relevance[:10]]) if sum(relevance[:10]) > 0 else 0.0

            results.append({
                'AUC': auc_score,
                'MRR': mrr_score,
                'nDCG@5': ndcg_5,
                'nDCG@10': ndcg_10,
            })

        except Exception:
            continue
    
    # Compute and return average metrics
    if results:
        results_df = pd.DataFrame(results)
        average_metrics = results_df.mean().to_dict()
        return average_metrics
    else:
        return {"AUC": 0.0, "MRR": 0.0, "nDCG@5": 0.0, "nDCG@10": 0.0}

# Example: Evaluate average metrics for a list of users
user_ids = dev_behaviors['userId'].unique()
average_metrics = evaluate_average_metrics_for_users(user_ids, get_similar_articles, dev_news, dev_behaviors, cosine_sim_tfidf)

# Print the result
print("Average Metrics Across Users:")
print(average_metrics)
```

```python
def evaluate_recommendations(user_id: str, recommender_func, news_df: pd.DataFrame, behaviors_df: pd.DataFrame, cosine_sim: np.ndarray, k: int = 10) -> Dict[str, float]:
    """Evaluate recommendations for a user."""
    user_behavior = behaviors_df[behaviors_df['userId'] == user_id]
    if user_behavior.empty or pd.isna(user_behavior['click_history'].iloc[0]):
        print(f"User ID {user_id} has no valid click history.")
        return None

    click_history = user_behavior['click_history'].iloc[0].split()
    if not click_history:
        print(f"User ID {user_id} has an empty click history after splitting.")
        return None

    try:
        # Use the latest clicked article's index if it is a valid integer index
        latest_click = click_history[-1]
        if latest_click not in news_df['itemId'].values:
            print(f"User ID {user_id} has an invalid latest click history item.")
            return None

        recommended_articles = recommender_func(latest_click, news_df, cosine_sim, num_recommendations=k)
        recommended_ids = recommended_articles['itemId'].tolist()
        relevance = [1 if item in click_history else 0 for item in recommended_ids]

        # Calculate AUC
        true_labels = [1 if item in click_history else 0 for item in news_df['itemId']]
        predicted_scores = [cosine_sim[news_df[news_df['itemId'] == latest_click].index[0], news_df[news_df['itemId'] == item].index[0]] if item in recommended_ids else 0 for item in news_df['itemId']]
        auc_score = roc_auc_score(true_labels, predicted_scores) if len(set(true_labels)) > 1 else 0.0

        # Calculate MRR
        mrr_score = calculate_mrr(relevance)

        # Calculate nDCG@5 and nDCG@10
        ideal_relevance = sorted(relevance, reverse=True)
        ndcg_5 = ndcg_score([ideal_relevance[:5]], [relevance[:5]]) if sum(relevance[:5]) > 0 else 0.0
        ndcg_10 = ndcg_score([ideal_relevance[:10]], [relevance[:10]]) if sum(relevance[:10]) > 0 else 0.0

        return {
            'AUC': auc_score,
            'MRR': mrr_score,
            'nDCG@5': ndcg_5,
            'nDCG@10': ndcg_10,
        }

    except Exception as e:
        print(f"Error occurred while getting recommendations for user {user_id}: {str(e)}")
        return None

# Example: Evaluate recommendations for a sample user
user_id = dev_behaviors['userId'].iloc[100]
metrics = evaluate_recommendations(user_id, get_similar_articles, dev_news, dev_behaviors, cosine_sim_tfidf)

if metrics:
    # Convert the metrics dictionary to a pandas DataFrame for table format
    metrics_df = pd.DataFrame.from_dict(metrics, orient='index', columns=['Score'])
    metrics_df.index.name = 'Metric'

    # Display the table using pandas
    print(metrics_df)
else:
    print(f"No valid recommendations for user {user_id}.")
```

### 7. Plot Metric Comparison

```python
from matplotlib.gridspec import GridSpec

def create_comparison_plots(tfidf_metrics: Dict[str, List[float]], 
                          llm_metrics: Dict[str, List[float]], 
                          save_path: str = None):
    """
    Create comprehensive visualization comparing TF-IDF and LLM metrics.
    
    Args:
        tfidf_metrics (dict): Dictionary containing TF-IDF metrics
        llm_metrics (dict): Dictionary containing LLM metrics
        save_path (str, optional): Path to save the plot
    """
    # Set up the style
    plt.style.use('default')  # Use default style instead of seaborn
    colors = ['#3361AC', '#E8AF30']  # Custom colors for better visualization
    
    # Create figure with GridSpec
    fig = plt.figure(figsize=(15, 10))
    gs = GridSpec(2, 2, figure=fig)
    
    # Box plot
    ax1 = fig.add_subplot(gs[0, 0])
    create_box_plot(tfidf_metrics, llm_metrics, ax1, colors)
    
    # Bar plot
    ax2 = fig.add_subplot(gs[0, 1])
    create_bar_plot(tfidf_metrics, llm_metrics, ax2, colors)
    
    # Add a super title
    fig.suptitle('Comparison of TF-IDF and LLM Metrics', 
                 fontsize=16, y=1.02)
    
    plt.tight_layout()
    
    if save_path:
        plt.savefig(save_path, dpi=300, bbox_inches='tight')
    
    return fig

def create_box_plot(tfidf_metrics: Dict[str, List[float]], 
                   llm_metrics: Dict[str, List[float]], 
                   ax: plt.Axes,
                   colors: List[str]):
    """Create box plot comparing metrics distribution."""
    metrics = ['AUC', 'MRR', 'nDCG@5', 'nDCG@10']
    data_to_plot = []
    labels = []
    
    for metric in metrics:
        data_to_plot.extend([tfidf_metrics[metric], llm_metrics[metric]])
        labels.extend([f'TF-IDF\n{metric}', f'LLM\n{metric}'])
    
    bp = ax.boxplot(data_to_plot, patch_artist=True)
    
    # Customize box colors
    for i, patch in enumerate(bp['boxes']):
        patch.set_facecolor(colors[i % 2])
        patch.set_alpha(0.7)
    
    # Customize whiskers, caps, and median lines
    for element in ['whiskers', 'caps']:
        for item in bp[element]:
            item.set_color('#2c3e50')
    
    for median in bp['medians']:
        median.set_color('#e74c3c')
        median.set_linewidth(2)
    
    ax.set_xticklabels(labels, rotation=45, ha='right')
    ax.set_ylabel('Score')
    ax.set_title('Distribution of Metrics')
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

def create_bar_plot(tfidf_metrics: Dict[str, List[float]], 
                   llm_metrics: Dict[str, List[float]], 
                   ax: plt.Axes,
                   colors: List[str]):
    """Create bar plot comparing mean metrics."""
    metrics = ['AUC', 'MRR', 'nDCG@5', 'nDCG@10']
    x = np.arange(len(metrics))
    width = 0.35
    
    tfidf_means = [np.mean(tfidf_metrics[m]) for m in metrics]
    llm_means = [np.mean(llm_metrics[m]) for m in metrics]
    
    # Add bars with error bars
    tfidf_std = [np.std(tfidf_metrics[m]) for m in metrics]
    llm_std = [np.std(llm_metrics[m]) for m in metrics]
    
    ax.bar(x - width/2, tfidf_means, width, label='TF-IDF', 
           color=colors[0], alpha=0.7, yerr=tfidf_std, capsize=5)
    ax.bar(x + width/2, llm_means, width, label='LLM', 
           color=colors[1], alpha=0.7, yerr=llm_std, capsize=5)
    
    ax.set_ylabel('Mean Score')
    ax.set_title('Mean Metric Comparison')
    ax.set_xticks(x)
    ax.set_xticklabels(metrics)
    ax.legend()
    ax.grid(True, alpha=0.3, linestyle='--')
    ax.spines['top'].set_visible(False)
    ax.spines['right'].set_visible(False)

def print_summary_statistics(tfidf_metrics: Dict[str, List[float]], 
                           llm_metrics: Dict[str, List[float]]):
    """Print summary statistics for metrics between TF-IDF and LLM models."""
    metrics = ['AUC', 'MRR', 'nDCG@5', 'nDCG@10']
    
    # Create a pretty formatted table
    print("\nSummary Statistics for TF-IDF and LLM Models:\n")
    print("-" * 80)
    print(f"{'Metric':<10} | {'Model':<8} | {'Mean':>10} | {'Median':>10} | {'Std Dev':>10}")
    print("-" * 80)
    
    for metric in metrics:
        tfidf_stats = {
            'mean': np.mean(tfidf_metrics[metric]) if tfidf_metrics[metric] else float('nan'),
            'median': np.median(tfidf_metrics[metric]) if tfidf_metrics[metric] else float('nan'),
            'std': np.std(tfidf_metrics[metric]) if tfidf_metrics[metric] else float('nan')
        }
        
        llm_stats = {
            'mean': np.mean(llm_metrics[metric]) if llm_metrics[metric] else float('nan'),
            'median': np.median(llm_metrics[metric]) if llm_metrics[metric] else float('nan'),
            'std': np.std(llm_metrics[metric]) if llm_metrics[metric] else float('nan')
        }
        
        print(f"{metric:<10} | {'TF-IDF':<8} | {tfidf_stats['mean']:>10.4f} | "
              f"{tfidf_stats['median']:>10.4f} | {tfidf_stats['std']:>10.4f}")
        print(f"{'':<10} | {'LLM':<8} | {llm_stats['mean']:>10.4f} | "
              f"{llm_stats['median']:>10.4f} | {llm_stats['std']:>10.4f}")
        print("-" * 80)

def collect_metrics(behaviors_df: pd.DataFrame, 
                   news_df: pd.DataFrame, 
                   tfidf_cosine_sim: np.ndarray, 
                   llm_cosine_sim: np.ndarray, 
                   num_users: int = 100):
    """Collect metrics for multiple users for both TF-IDF and LLM similarity."""
    tfidf_metrics = {'AUC': [], 'MRR': [], 'nDCG@5': [], 'nDCG@10': []}
    llm_metrics = {'AUC': [], 'MRR': [], 'nDCG@5': [], 'nDCG@10': []}
    
    for i in range(min(num_users, len(behaviors_df))):
        user_id = behaviors_df['userId'].iloc[i]
        
        # Evaluate TF-IDF model for the user
        tfidf_user_metrics = evaluate_recommendations(user_id, get_similar_articles, 
                                                    news_df, behaviors_df, tfidf_cosine_sim)
        if tfidf_user_metrics:
            for key in tfidf_metrics:
                tfidf_metrics[key].append(tfidf_user_metrics[key])
        
        # Evaluate LLM model for the user
        llm_user_metrics = evaluate_recommendations(user_id, get_similar_articles, 
                                                  news_df, behaviors_df, llm_cosine_sim)
        if llm_user_metrics:
            for key in llm_metrics:
                llm_metrics[key].append(llm_user_metrics[key])
    
    return tfidf_metrics, llm_metrics

# Main execution
try:
    # Load precomputed data
    vectorizer, dev_tfidf, dev_embeddings, cosine_sim_tfidf, cosine_sim_llm, dev_news = load_precomputed_data()
    
    # Collect metrics for a subset of users
    tfidf_metrics, llm_metrics = collect_metrics(dev_behaviors, dev_news, 
                                               cosine_sim_tfidf, cosine_sim_llm, 
                                               num_users=100)
    
    # Print summary statistics
    print_summary_statistics(tfidf_metrics, llm_metrics)
    
    # Create and display visualizations
    fig = create_comparison_plots(tfidf_metrics, llm_metrics, save_path='metrics_comparison.png')
    plt.show()
    
except Exception as e:
    print(f"An error occurred: {e}")
```
